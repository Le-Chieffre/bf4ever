```python
import numpy as np
import pandas as pd
from pandas import DataFrame, Series

NPI_yr = pd.read_excel("./NPI (yr change).xlsx")

NPI_yr.describe()
NPI_yr.drop(NPI_yr.columns[0],axis=1,inplace=True)

NPI_yr.rename(columns={NPI_yr.columns[0]:"Period",NPI_yr.columns[1]:"YoY Return"},inplace=True)

NPI_yr.dropna(inplace=True)

def fuck_dot(row) -> bool:
    return row["YoY Return"] == ".."

mapper = NPI_yr.apply(fuck_dot, axis=1)
NPI_yr.drop(NPI_yr[mapper].index, inplace=True)
```

```python
# Re-calc to per cent
def re_calc(row):
    return row["YoY Return"]/100
NPI_yr["YoY in %"] = NPI_yr.apply(re_calc, axis=1)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(30,10))
plt.plot(NPI_yr["Period"], NPI_yr["YoY in %"])

plt.xticks(NPI_yr["Period"][::12])

plt.xlabel("Period")
plt.ylabel("YoY in %")
plt.show()
```
